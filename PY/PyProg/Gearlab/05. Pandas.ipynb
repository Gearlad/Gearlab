```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize = (6, 6))
plt.xkcd()

plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.legend('lower left')

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xlim(0, 10)
plt.ylim(-1, 1)

plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(-1, 1, 0.5))

plt.title('Sin and Cos')
```

```python
def ARGV(*argv):
    for i in argv:
        print(i)
```

```python
ARGV("Hello", "Hi")
```

```python
def ARGV1(arg, *argv):
    print(arg, "\n")
    for i in argv:
        print(i)
```

```python
ARGV1("my", "name", "is", "snyder")
```

```python
def KWARG(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} == {value}")
```

```python
KWARG(first = "N", second = "C", third = "T", fourth = "S")
```

```python
def Four(arg1, arg2, arg3, arg4):
    print("arg1 is ", arg1)
    print("arg2 is ", arg2)
    print("arg3 is ", arg3)
    print("arg4 is ", arg4)
```

```python
args = ("N", "T", "C", "S")
```

```python
Four(*args)
```

```python
kwargs = {"arg2" : "T", "arg1" : "N", "arg4": "S", "arg3" : "C"}
```

```python
Four(**kwargs)
```
