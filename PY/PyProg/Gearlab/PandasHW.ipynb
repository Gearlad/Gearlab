```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib as mpl
```

```python
df = pd.read_csv('eth.csv')
#data ranging from 2015/8/7 to present
df = df[8:]
df.head()
```

```python
#Background: blocks on Ethereum are generated on average once every twelve seconds. The average calculation difficulty
#in creating each block for a single computer has increased enormously since Ethereum's inception. This graph provides
#an analysis in the variance of block generation difficulty from 2015 to present.

difficulty = df.loc[:, 'averageDifficulty']
difficulty
```

```python
date = df.loc[:, 'date']
date
```

```python
font = {'family': 'sans-serif',
        'weight': 'bold',
       'size'   : 22}
mpl.rc('font', **font)
fig, ax = plt.subplots(figsize = (20,10))

plt.plot(difficulty)
plt.xlabel("Days Since Inception [2015/08/07] - [2019-03-25]")
plt.ylabel("Mining Difficulty Level")
```
